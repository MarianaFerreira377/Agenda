```python
import pandas as pd


df = pd.read_csv("https://raw.githubusercontent.com/eduardapicolo/PI5-turma01-Grupo03/feature/data-encoding/data/animais_adotar.csv")

df=df[["tipo_animal","sexo","porte","idade","pelagem","cuidados_veterinarios","vive_bem_com"]].sample(n=20)


df["idade"] = df["idade"].astype(str).str.strip().str.lower()

df = df[["tipo_animal","sexo","porte","idade","pelagem","cuidados_veterinarios","vive_bem_com"]]

for col in df.columns:
    df[col] = df[col].astype(str).str.strip().str.lower()

df = pd.get_dummies(df, columns=df.columns, dtype=int)

# salvar
df.to_csv("dados_tratados.csv", sep=";", index=False)
```

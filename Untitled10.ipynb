```python
#Questão 1
i1, i2, i3, i4 = 0, 0, 0, 0

while True:
    num = float(input("Digite um número (negativo para sair): "))
    if num < 0:
        break
    if 0 <= num <= 25:
        i1 += 1
    elif 26 <= num <= 50:
        i2 += 1
    elif 51 <= num <= 75:
        i3 += 1
    elif 76 <= num <= 100:
        i4 += 1

print(f"[0-25]: {i1} | [26-50]: {i2} | [51-75]: {i3} | [76-100]: {i4}")
```

```python
#Questão 2
pop_a = 1000000  # País A: 1 milhão [cite: 6]
pop_b = 500000   # País B: 500 mil [cite: 6]
anos = 0

while pop_b <= pop_a:
    pop_a += pop_a * 0.05  # Crescimento de 5% [cite: 7]
    pop_b += pop_b * 0.20  # Crescimento de 20% [cite: 7]
    anos += 1
    print(f"Ano {anos}: Pop A = {pop_a:.0f}, Pop B = {pop_b:.0f}")

print(f"\nSerão necessários {anos} anos para que a população B supere a A.")
```

```python
# Questão 3
p = float(input("Digite o peso em kg: "))
h = float(input("Digite a altura em metros: "))

imc = p / (h ** 2)

print(f"Resultado do IMC: {imc:.2f}")

if imc < 20:
    print("Abaixo do Normal")
elif 20 <= imc < 25:
    print("Normal")
elif 25 <= imc < 30:
    print("Sobrepeso")
elif 30 <= imc < 35:
    print("Obesidade leve")
elif 35 <= imc < 40:
    print("Obesidade moderada")
else:
    print("Obesidade mórbida")
```

```python
#Questão 6
def calcular_atraso(valor, dias):
    # Juros de 10% por dia de atraso conforme enunciado
    novo_valor = valor + (valor * 0.10 * dias)
    return novo_valor

v = float(input("Valor da parcela: "))
d = int(input("Dias de atraso: "))
print(f"Valor atualizado: R$ {calcular_atraso(v, d):.2f}")
```

```python
# Questão 7

def realizar_conversao(celsius, escala):
    if escala == 'K':
        resultado = celsius + 273
        return resultado
    elif escala == 'F':
        resultado = (1.8 * celsius) + 32
        return resultado
    else:
        return None

print("--- Conversor de Temperatura ---")
temp_c = float(input("Digite a temperatura em Celsius: "))

print("Escolha a escala para conversão:")
print("[K] para Kelvin")
print("[F] para Fahrenheit")
opcao = input("Sua opção: ").upper()

valor_convertido = realizar_conversao(temp_c, opcao)

if valor_convertido is not None:
    print(f"\nA temperatura convertida é: {valor_convertido:.2f}")
else:
    print("\nErro: Escala inválida selecionada.")
```

```python
# Questão 8
notas = []
while True:
    n = float(input("Digite uma nota (-1 para encerrar): "))
    if n == -1:
        break
    notas.append(n)

# a) Quantidade de valores lidos
qtd = len(notas)
print(f"\na) Quantidade de valores lidos: {qtd}")

# b) Valores na ordem em que foram informados (lado a lado)
print("b) Valores informados: ", end="")
print(*notas, sep=" ")

# c) Valores na ordem inversa (um abaixo do outro)
print("c) Valores na ordem inversa:")
for n in reversed(notas):
    print(n)

# d) Soma dos valores
soma = sum(notas)
print(f"d) Soma dos valores: {soma}")

# e) Média dos valores
media = soma / qtd if qtd > 0 else 0
print(f"e) Média dos valores: {media:.2f}")

# f) Quantidade de valores acima da média calculada
acima_media = len([n for n in notas if n > media])
print(f"f) Quantidade de valores acima da média: {acima_media}")

# g) Quantidade de valores abaixo de sete
abaixo_sete = len([n for n in notas if n < 7])
print(f"g) Quantidade de valores abaixo de sete: {abaixo_sete}")

# h) Encerramento
print("h) Programa Finalizado!")
```

```python
#Questão 9
def transformar_lista(lista):
    for i in range(len(lista)):
        if i % 2 == 0:
            lista[i] = lista[i] * 3
        else:
            lista[i] = lista[i] ** 2
    return lista

def ordenar_especifico(lista):
    parcial1 = sorted(lista[:5], reverse=True)
    parcial2 = sorted(lista[5:], reverse=True)
    return parcial1 + parcial2

numeros = []
for i in range(10):
    numeros.append(float(input(f"Digite o {i+1}º número: ")))

lista_transformada = transformar_lista(numeros)
print(f"Lista transformada: {lista_transformada}")

lista_ordenada = ordenar_especifico(lista_transformada)
print(f"Lista ordenada exemplo: {lista_ordenada}")
```

```python
#Questão 10
import math

def resolver_equacao_segundo_grau():
    a = float(input("Digite o valor de a: "))

    if a == 0:
        print("A equação não é do segundo grau. Programa encerrado.")
        return

    b = float(input("Digite o valor de b: "))
    c = float(input("Digite o valor de c: "))

    delta = (b**2) - (4*a*c)

    if delta < 0:
        print(f"Delta = {delta}. A equação não possui raízes reais. Programa encerrado.")
    elif delta == 0:
        raiz = -b / (2 * a)
        print(f"Delta = 0. A equação possui apenas uma raiz real: {raiz}")
    else:
        x1 = (-b + math.sqrt(delta)) / (2 * a)
        x2 = (-b - math.sqrt(delta)) / (2 * a)
        print(f"Delta = {delta}. A equação possui duas raízes reais:")
        print(f"x1 = {x1}")
        print(f"x2 = {x2}")

resolver_equacao_segundo_grau()
```

```python
#Questão 12
votos = {'O': 0, 'B': 0, 'R': 0, 'U': 0}
total = 0

while True:
    voto = input("Opinião (O/B/R/U) ou 'S' para sair: ").upper()
    if voto == 'S':
        break
    if voto in votos:
        votos[voto] += 1
        total += 1
    else:
        print("Opção inválida.")

if total > 0:
    for opcao, qtd in votos.items():
        porcentagem = (qtd / total) * 100
        print(f"Opção {opcao}: {porcentagem:.2f}%")
```

```python
#Questão 13
import math

def calcular_fatoriais():
    numeros = []
    for i in range(10):
        numeros.append(int(input(f"Digite o {i+1}º número inteiro: ")))

    print("\nResultados dos fatoriais:")
    for n in numeros:
        res = math.factorial(n)
        print(f"{n}! = {res}")

calcular_fatoriais()
```

```python
#Questão 14
import numpy as np

def Estatisticas(lista):
    estatisticas = {
        "Média": np.mean(lista),
        "Mediana": np.median(lista),
        "Variância": np.var(lista),
        "Desvio Padrão": np.std(lista),
        "Maior valor": np.max(lista),
        "Menor valor": np.min(lista)
    }
    return estatisticas

# Programa Principal
numeros = []

while True:
    try:
        n = float(input("Digite um número (-1 para encerrar): "))
        if n == -1:
            break
        numeros.append(n)
    except ValueError:
        print("Entrada inválida. Digite um número.")

if len(numeros) > 0:
    resultados = Estatisticas(numeros)
    print("\n--- Estatísticas da Lista ---")
    for chave, valor in resultados.items():
        print(f"{chave}: {valor:.2f}")
else:
    print("Nenhum valor foi inserido na lista.")
```
